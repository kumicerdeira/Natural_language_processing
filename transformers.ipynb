```python
from transformers import pipeline

# 質疑応答
nlp = pipeline("question-answering")

context = r"""
Extractive Question Answering is the task of extracting an answer from a text given a question. An example of a
question answering dataset is the SQuAD dataset, which is entirely based on that task. If you would like to fine-tune
a model on a SQuAD task, you may leverage the `run_squad.py`.
"""

print(nlp(question="What is extractive question answering?", context=context))
#返ってきた応答を出力
```

```python
#BERTとは、Bidirectional Encoder Representations from Transformers「Transformerによる双方向のエンコード表現」
#自然言語処理モデル
!pip install fugashi ipadic
#日本語対応のためのライブラリ

from transformers import pipeline

classifier = pipeline(
    task="sentiment-analysis",
    model="koheiduck/bert-japanese-finetuned-sentiment",
    tokenizer="koheiduck/bert-japanese-finetuned-sentiment"
)

sentence1 = "嬉しい"
sentence2 = "悲しい"
result1 = classifier(sentence1)[0]
result2 = classifier(sentence2)[0]
print(f"sentence1: {sentence1}, label: {result1['label']}, with score: {round(result1['score'], 4)}")
print(f"sentence2: {sentence2}, label: {result2['label']}, with score: {round(result2['score'], 4)}")
```

```python
#スペイン語モデル
from transformers import AutoModel, AutoTokenizer

# スペイン語BERTモデルの名前
model_name = "dccuchile/bert-base-spanish-wwm-cased"

# モデルとトークナイザーのロード
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
# テキストのトークン化
text = "¡Hola, mundo!"
inputs = tokenizer(text, return_tensors="pt")

# モデルによる推論
outputs = model(**inputs)
```

```python
from transformers import pipeline

# スペイン語の感情分析モデルの設定
classifier = pipeline(
    task="sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment",  # スペイン語に対応したモデル
    tokenizer="nlptown/bert-base-multilingual-uncased-sentiment"
)

# スペイン語の文
sentence1 = "bien"
sentence2 = "mal"

# 感情分析の実行
result1 = classifier(sentence1)[0]
result2 = classifier(sentence2)[0]

# 結果の表示
print(f"sentence1: {sentence1}, label: {result1['label']}, with score: {round(result1['score'], 4)}")
print(f"sentence2: {sentence2}, label: {result2['label']}, with score: {round(result2['score'], 4)}")
```
